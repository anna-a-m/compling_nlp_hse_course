```python
def remove_tags(text: str):
    # remove tags
    new_text = re.sub(r'\w+=".+?"', '', text)   # убираем все аттрибуты тегов
    new_text = re.sub(r'(<\/?\w+\s*>)', ' ', new_text) # убираем все теги
    new_text = re.sub(r' +', ' ', new_text)   # убираем лишние пробелы
    new_text = re.sub(r' \.', '.', new_text)   # убираем пробелы перед точками
    new_text = re.sub(r'([а-яё])\s\n', r'\1.\n', new_text)   # добавляем точку в предложения в конце строки без нее
    return new_text.strip()   # возвращаем текст без пробелов в начале/в конце
```

```python
mystem_lines = [json.loads(line) for line in open('text_parsed.txt', encoding='utf-8')]
```

```python
# генератор, чтобы делить выделенные токены по концу абзаца или предложения
def funny_gen(token_list: list):
    one_sent = []
    for t in token_list:
        if t not in ["\\s", " \r\n", "\r\n"]:
            one_sent.append(t)
        else:
            yield one_sent
            one_sent = []
    yield one_sent
```

```python
# находим все токены
ts = [d['text'] for ml in mystem_lines for d in ml]
```

```python
mystem_sentences = list(funny_gen(ts))
```

```python
# пример поделенного предложения
mystem_sentences[2]
```

```python
# поделим текст на предложения через razdel.sentenize
razdel_sentences = list(sentenize(TEXT))
```

```python
# по длине они сравнимы
len(razdel_sentences), len(mystem_sentences)
```

**Попробуем теперь последовательно сопоставить получившиеся предложения**

```python
for i in range(len(mystem_sentences)):
    if razdel_sentences[i].text.strip() != ''.join(mystem_sentences[i]).strip():
        print('R\t', razdel_sentences[i].text)
        print('---')
        print('M\t', ''.join(mystem_sentences[i]))
        print('===')
```
